```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

# File to Load
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
city_data_df.head()
```

```python
# Combine the data into a single dataset
combined_data_df = pd.merge(city_data_df, ride_data_df, how='outer', on='city')
# Display the data table for preview
combined_data_df.head()
```

```python
combined_data_df = combined_data_df[['city', 'date', 'fare', 'ride_id', 'driver_count', 'type']]
combined_data_df['type'].value_counts()
```

## Bubble Plot of Ride Sharing Data

```python
combined_data_rural = combined_data_df.loc[combined_data_df['type'] =='Rural']
combined_data_urban = combined_data_df.loc[combined_data_df['type'] =='Urban']
combined_data_suburban = combined_data_df.loc[combined_data_df['type'] =='Suburban']

```

```python
# Obtain the x and y coordinates for each of the three city types
rural_fare = combined_data_rural.groupby(['city'])['fare'].mean()
rural_num_rides = combined_data_rural['city'].value_counts()

sub_fare = combined_data_suburban.groupby(['city'])['fare'].mean()
sub_num_rides = combined_data_suburban['city'].value_counts()

urb_fare = combined_data_urban.groupby(['city'])['fare'].mean()
urb_num_rides = combined_data_urban['city'].value_counts()

# #Set variable for number of drivers for each type, which will determine bubble size
rural_num_drivers = combined_data_rural.groupby(['city'])['driver_count'].mean()
sub_num_drivers = combined_data_suburban.groupby(['city'])['driver_count'].mean()
urb_num_drivers = combined_data_urban.groupby(['city'])['driver_count'].mean()


# # Build the scatter plots for each city types
plt.scatter(rural_num_rides, rural_fare, marker="o", color='y', edgecolors='black', s=rural_num_drivers*5, alpha=.9)
plt.scatter(sub_num_rides, sub_fare, marker="o", color='b', edgecolors='black', s=sub_num_drivers*5, alpha=.50)
plt.scatter(urb_num_rides, urb_fare, marker="o", color='r', edgecolors='black', s=urb_num_drivers*5, alpha=.50)
# Incorporate the other graph properties
plt.grid(which='major', axis='both')
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')

# Create a legend
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Urban',
                          markerfacecolor='r', markersize=8, alpha=.75),
                   Line2D([0], [0], marker='o', color='w', label='Suburban',
                          markerfacecolor='b', markersize=8, alpha=.75),
                  Line2D([0], [0], marker='o', color='w', label='Rural',
                          markerfacecolor='y', markersize=8, alpha=.75),]

plt.legend(handles=legend_elements, title = 'City Types', loc='best', fancybox = True)


# Incorporate a text label regarding circle size
# note = 'Circle size correlates with driver count per city.'
plt.text(42.5, 35, 'Note: \nCircle size correlates with driver count per city.')

# Save Figure
plt.savefig("Data/Pyber_Ride_Sharing_Data_2016.png")
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_pct = round(combined_data_urban['fare'].sum() / combined_data_df['fare'].sum(), 3) *100
rural_pct = round(combined_data_rural['fare'].sum() / combined_data_df['fare'].sum(), 3) *100
sub_pct = round(combined_data_suburban['fare'].sum() / combined_data_df['fare'].sum(), 3) *100

# Build Pie Chart
labels = ["Urban", "Rual", "Suburban"]
sizes = [urban_pct, rural_pct, sub_pct]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.title('% of Total Fares by City Type')

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=-75)

# plt.title('Pyber Ride Sharing Data (2016)')

# Save Figure
plt.savefig("Data/Percent_of_Total_Fares_by_City_Type.png")
plt.show
```
